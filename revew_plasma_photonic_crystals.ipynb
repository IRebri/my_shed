```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import math
tau_p=8.5*(10**(-12)) # s
w_0 = 2*math.pi*55*(10**9) # Hz
E_0 = 10**6 # V/m
t1 = np.linspace(0,2.25*(10**(-11)),100) # why 10**(-9) everything is settled earlyer after 10**(-11)
t2 = np.linspace(0,2.25*(10**(-9)),1000)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.subplot(141)
plt.plot(t1, E_0*(0.5+np.cos(w_0*t1)+np.cos(2*w_0*t1))*np.exp(-(t1/tau_p)**2))
plt.xlabel(r'\textbf{time} (s)')
plt.ylabel(r'\textit{Electric field} (V/m)',fontsize=16)
plt.title(r"$\displaystyle\ ~10^{-11}sec$", fontsize=16)

plt.subplot(144)
plt.plot(t2, E_0*(0.5+np.cos(w_0*t2)+np.cos(2*w_0*t2))*np.exp(-(t2/tau_p)**2))
plt.xlabel(r'\textbf{time} (s)')
plt.xlim([-0.1*(10**(-9)), 2.5*(10**(-9))])
plt.ylabel(r'\textit{Electric field} (V/m)',fontsize=16)

plt.title(r"$\displaystyle\ ~10^{-9}sec$", fontsize=16)

# Make room for the ridiculously large title.
plt.subplots_adjust(top=0.8)



# Make room for the ridiculously large title.
plt.subplots_adjust(top=0.8)
```
